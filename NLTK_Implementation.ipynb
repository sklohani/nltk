### Word Replacement

```python
import re
from nltk.corpus import wordnet
```

```python
R_patterns = [
    (r'won\'t', 'will not'),
    (r'can\'t', 'cannot'),
    (r'i\'m', 'i am'),
    (r'(\w+)\'ll', '\g<1> will'),
    (r'(\w+)n\'t', '\g<1> not'),
    (r'(\w+)\'ve', '\g<1> have'),
    (r'(\w+)n\'s', '\g<1> is'),
    (r'(\w+)n\'re', '\g<1> are')
]
```

```python
class REReplacer(object):
    def __init__(self, patterns = R_patterns):
        self.patterns = [(re.compile(regex), repl) for (regex, repl) in patterns]
    
    def replace(self, text):
        s = text
        for (pattern, repl) in self.patterns:
            s = re.sub(pattern, repl, s)
        return s
```

```python
rep_word = REReplacer()
rep_word = REReplacer()
rep_word.replace("I won't do it")
```

```python
rep_word = REReplacer()
for token in mothersLegacyToken:
    print(rep_word.replace(token), end=" ")
```

```python
rep_word = REReplacer()
for token in proudPrincessToken:
    print(rep_word.replace(token), end=" ")
```

As we can see from the outputs of above cells, using the custom class for replacing the words having punctuations or contraction words are effectively replaced with the corresponding word.

```python
class Rep_word_remover(object):
    def __init__(self):
        self.repeat_regex = re.compile(r'(\w*)(\w)\2(\w*)')
        self.repl = r'\1\2\3'
    def replace(self, word):
        if wordnet.synsets(word):
            return word
        replace_word = self.repeat_regex.sub(self.repl, word)
        if replace_word != word:
            return self.replace(replace_word)
        else:
            return replace_word
```

```python
rep_word = Rep_word_remover()
rep_word.replace("Hiiii")
```

The user-defined class `Rep_word_remover` is replacing the word with repetiton characters with more than equal to 2 repetiton characters. As seen in above output it replaced the unneccessary characters.

As the input files does not have any word with repetiton characters so this operation is irrelevent for these input files.

```python
rep_word = Rep_word_remover()
for token in proudPrincessToken:
    print(rep_word.replace(token), end = " ")
```

```python
class Word_syn_replacer(object):
    def __init__(self, word_map):
        self.word_map = word_map
    def replace(self, word):
        return self.word_map.get(word, word)
```

```python
rep_syn = Word_syn_replacer({"GN": "Good Night"})
rep_syn.replace("GN")
```

The user-defined class `Word_syn_replacer` is replacing the associated word with its common synonym.

```python
class word_antonym_replacer(object):
    def replace(self, word, pos=None):
        antonyms = set()
        for syn in wordnet.synsets(word, pos=pos):
            for lemma in syn.lemmas():
                for antonym in lemma.antonyms():
                    antonyms.add(antonym.name())
        if len(antonyms) == 1:
            return antonyms.pop()
        else:
            return "_"
        
    def replace_negations(self, sent):
        i, l = 0, len(sent)
        words = []
        while i<l:
            word = sent[i]
            if word == 'not' and i+1 < l:
                ant = self.replace(sent[i+1])
                if ant:
                    words.append(ant)
                    i+=2
                    continue
            words.append(word)
            i+=1
        return words
```

```python
rep_antonym = word_antonym_replacer()
rep_antonym.replace("uglify")
```

```python
sentence = ["Let's", "not", "demoralize", "our", "country"]
rep_antonym = word_antonym_replacer()
rep_antonym.replace_negations(sentence)
```

```python
rep_antonym = word_antonym_replacer()
for token in mothersLegacyToken:
    print(rep_antonym.replace(token), end = " ")
```

```python
rep_antonym = word_antonym_replacer()
for token in proudPrincessToken:
    print(rep_antonym.replace(token), end = " ")
```

The user-defined `word_antonym_replacer` helps in replacing the words with its antonyms whose synonym can be found in `synsets` of `wordnet` module. As seen from the above cells the word which can be find in synsets of wordnet is replaced with its antonyms.
